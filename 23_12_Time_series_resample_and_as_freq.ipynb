```
from pandas_datareader import data 
#using the google data 
goog = data.DataReader("GOOG",start="2000",end="2020",data_source="yahoo")
#using just the closing price
goog = goog["Close"]
#plotting
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.figure(figsize=(10,5))
goog.plot()
```

```
plt.figure(figsize=(10,5))
#downsampling 
goog.plot(alpha=0.5,style="-")
#resample method
goog.resample("BA").mean().plot(style="-")
#asfreq() method
goog.asfreq("BA").plot(style="-")
plt.legend(["input","resample","asfreq"],loc="upper left")
```

the resample() method returns the avg of the prev period 
while the asfreq() method returns the value at the brgining of the period
